```python
def check_manifest(manifest_file):
    with open(manifest_file, 'r') as f:
        count = 0
        for line in f:
            count += 1
            entry = json.loads(line)
            # Check if the file exists
            if not os.path.exists(entry['audio_filepath']):
                print(f"Missing file: {entry['audio_filepath']}")
            # Check if the text label is valid
            if entry['text'] not in vocab:
                print(f"Invalid label: {entry['text']}")
        print(f"Total files in {manifest_file}: {count}")

check_manifest(train_manifest)
check_manifest(val_manifest)
```

```python
def create_manifest(dataset_dir, output_path):
    with open(output_path, 'w') as f_out:
        for label_dir in os.listdir(dataset_dir):
            if os.path.isdir(os.path.join(dataset_dir, label_dir)):
                for wav_file in os.listdir(os.path.join(dataset_dir, label_dir)):
                    if wav_file.endswith('.wav'):
                        entry = {
                            'audio_filepath': os.path.join(dataset_dir, label_dir, wav_file),
                            'duration': librosa.core.get_duration(path=os.path.join(dataset_dir, label_dir, wav_file)),
                            'text': label_dir
                        }
                        f_out.write(json.dumps(entry) + '\n')
```

```python
total_list = []
for label in tqdm(os.listdir('./data/speech_commands')):
  if not os.path.isdir(os.path.join('./data/speech_commands', label)):
    continue
  if (label == "_background_noise_"):
    continue
  for fname in os.listdir(os.path.join('./data/speech_commands', label)):
    if fname.endswith('.wav'):
      total_list.append(f"./data/speech_commands/{label}/{fname}")
```

```python
testing_list = []
validation_list = []
with open('./data/speech_commands/testing_list.txt', 'r') as fin:
  for line in fin:
    testing_list.append('./data/speech_commands/' + line.rstrip())
with open('./data/speech_commands/validation_list.txt', 'r') as fin:
  for line in fin:
    validation_list.append('./data/speech_commands/' + line.rstrip())

training_list = []
for i in tqdm(total_list):
  if (i not in testing_list) and (i not in validation_list):
    training_list.append(i)
```

```python
from tqdm.notebook import tqdm
```

```python
import json


def build_manifest(dataList, manifest_path):
  with open(manifest_path, 'w') as fout:
    for dataPath in tqdm(dataList):
      transcript = dataPath.split('/')[-2]
      duration = librosa.core.get_duration(path=dataPath)
      metadata = {
                    "audio_filepath": dataPath,
                    "duration": duration,
                    "text": transcript
                }
      json.dump(metadata, fout)
      fout.write('\n')

# Building Manifests
print("******")
train_manifest = './train_manifest.json'
build_manifest(training_list, train_manifest)
print("Training manifest created.")

test_manifest = './test_manifest.json'
build_manifest(testing_list, test_manifest)
print("Test manifest created.")
print("***Done***")
```
